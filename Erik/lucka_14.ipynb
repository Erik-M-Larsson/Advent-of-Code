```python
långfil = "input_data/14.txt"
with open(långfil, "r") as f:
    kartdata = f.read().splitlines()
```

```python
def fritt_fall(r, k, karta):
    if karta[r, k]:
        return True
    
    fallväg = karta[r, k:]
    try:
        k = k + np.where(fallväg)[0][0]
    except IndexError:
        return True

    if not karta[r-1, k]:
        if fritt_fall(r-1, k, karta):
            return True
    elif not karta[r+1, k]:
        if fritt_fall(r+1, k, karta):
            return True
    else:
        karta[r, k-1] = 2
    
    return False
```

```python
def rita(karta):
    clear_output(wait=True)
    
    for r in karta.T:
        for k in r:
            print('▄' if k==1 else 'o' if k==2 else'.', end=" ")
        print("")
```

```python
lägsta_punkt = max([pkt[1] for rad in kartdata for pkt in eval('(' + rad.replace(' -> ', '),(') + ')')])

x_positioner = [pkt[0] for rad in kartdata for pkt in eval('(' + rad.replace(' -> ', '),(') + ')')]
x_min, x_max = min(x_positioner), max(x_positioner)
```

```python
bredd_karta = 400
höjd_karta = 200
x_start = 500 - bredd_karta//2

karta = np.zeros((bredd_karta,höjd_karta),int)

for rad in kartdata:
    föregående_pkt = None
    for pkt in eval('(' + rad.replace(' -> ', '),(') + ')'):
        if föregående_pkt:
            r = (föregående_pkt[0]-x_start, pkt[0]-x_start)
            k = (föregående_pkt[1], pkt[1])
        
            karta[min(r):max(r)+1, min(k):max(k)+1]= 1
        föregående_pkt = pkt

if deluppgift == 2:
    karta[:, lägsta_punkt + 2] = 1

rita(karta[x_min-x_start-2:x_max-x_start+3, 0:lägsta_punkt+3])
```

```python
fullt = False
while not fullt:
    fullt = fritt_fall(500-x_start, 0, karta)

rita(karta[x_min-x_start-2:x_max-x_start+3, 0:lägsta_punkt+3])

print("Svar:", np.where(karta==2)[1].shape[0])
```
